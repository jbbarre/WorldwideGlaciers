```python
import cmasher as cmr
import matplotlib.colors as colors
from matplotlib.colors import LinearSegmentedColormap
import numpy as np
```

```python
# Used to reconstruct the colormap in viscm
parameters = {'xp': [22.674387857633945, 11.221508276482126, -14.356589454756971, -47.188177587392218, -34.590010048125208, 0.15039134803535603],
              'yp': [-20.102530541012214, -33.082460732984288, -42.24476439790574, -5.5955497382198871, 42.5065445026178, 24.563699825479944],
              'min_JK': 18.8671875,
              'max_JK': 92.5}
```

```python
# Construct an RGB table using a log scale between 1 and 3000 m/year.
vel = np.exp(np.linspace(np.log(1), np.log(3000), num=256)) 
hue = np.arange(256) / 255.0 
sat = np.clip(1. / 3 + vel / 187.5, 0, 1) 
value = np.zeros(256) + 0.75 
hsv = np.stack((hue, sat, value), axis=1) 
rgb = colors.hsv_to_rgb(hsv) 
# Be sure the first color (the background) is white
rgb[0, :] = 1
# Create the colormap 
cmap = colors.ListedColormap(rgb, name='velocity')
```

```python

cm_data = cmr.take_cmap_colors(cmap, 256)
np.asarray(cm_data)
```
